## 6.1. Medidas de tendencia central
Las medidas de tendencia central son herramientas estadísticas que permiten identificar el valor representativo o central de un conjunto de datos. Estas medidas resumen la información mediante un único valor, facilitando la comprensión del comportamiento general de una variable.

En esta sección se calcularán la media aritmética, la mediana y la moda del precio de los computadores portátiles, con el propósito de obtener una primera descripción de la distribución de los datos.

### 6.1.1 Media aritmética

La media aritmética representa el valor promedio de un conjunto de datos. Se obtiene sumando todos los valores de la variable y dividiendo el resultado entre el número total de observaciones.

En este análisis se calculará la media del precio de los computadores portátiles.

```python
# Calculando media de precio de computadores portatiles
media_precio=round(np.mean(df["price"]),2)
print(f"La media del precio de los computadores es: {media_precio}")
```

### 6.1.2 Mediana

La mediana corresponde al valor central de un conjunto de datos ordenado de menor a mayor. A diferencia de la media, la mediana es menos sensible a valores extremadamente altos o bajos, por lo que representa mejor el valor central cuando existen datos atípicos.

En este análisis se calculará la mediana del precio de los computadores portátiles.

```python
# Calculando la Mediana de precios de laptops con Numpy
mediana_precio=round(np.median(df["price"]),2)
print(f"La mediana del precio de los computadores es: {mediana_precio}")
```

Otra forma de calcularlo con Stats.

```python
#Calculando la Mediana con Stats.
mediana_precio2=stats.median(df["price"])
print(f"La mediana del precio de los computadores es: {mediana_precio2}")
#
```

Tambien existe la posibilidad de calcularlo con Pandas

```python
#Calculando la Mediana con función de Pandas
mediana_precio = round(df["price"].median(), 2)
print(f"La mediana del precio de los computadores es: {mediana_precio}")
```

#### **Interpretación:**
La **mediana** del precio es 61.990, mientras que la media es 79.907,41. Dado que la media es considerablemente mayor que la **mediana**, es probable que existan computadores de precios muy elevados que incrementan el promedio del conjunto de datos.

### 6.1.3 Moda

La moda es el valor o dato que más veces se repite en un grupo.

```python
#Calculando la Moda
moda_precio=stats.mode(df["price"])
print(f"El precio que mas se repite es: {moda_precio}")
```

```python
#calculamndo Moda con Pandas
moda_preciopd = df["price"].mode()
print(f"El precio que mas se repite es: {moda_preciopd}")
```

```python
#Solo muestra la moda que este en el espacio 0
moda_preciopd2 = df["price"].mode()[0]
print(moda_preciopd2)
```

**Para concluir,** las **medidas de tendencia central** en este conjunto de datos serian las siguientes:

|Medida	   |Valor     |
|----------|----------|
|Media	   |79907,41  |
|Mediana   |61990     |
|Moda	     |49990     |

La media (79.907,41) es superior a la mediana (61.990) y a la moda (49.990). Esto sugiere que existen computadores de precio elevado que incrementan el valor promedio del conjunto de datos, generando una distribución con **asimetría hacia la derecha.**

Para determinar si los datos presentan un comportamiento homogéneo o heterogéneo será necesario analizar las medidas de dispersión en la siguiente sección.

##6.2.Medidas de dispersión

Las medidas de dispersión para datos agrupados (como el rango, la varianza, la desviación estándar y el coeficiente de variación) cuantifican qué tan alejados están los datos del valor central.
Las medidas de dispersión muestran qué tan separados o agrupados están los datos respecto a su valor central.

###6.2.1 Rango

El rango es la medida de dispersión más sencilla. Se obtiene restando el valor mínimo al valor máximo de un conjunto de datos y permite conocer la amplitud total de los valores observados.

Rango = Límite superior final - Límite inferior inicial

```python
#Calculando el Rango de precios
precio_max = df["price"].max()
print(f"El precio maximo es: {precio_max}")
precio_min = df["price"].min()
print(f"El precio minimo es: {precio_min}")
rango_price= precio_max - precio_min
print(f"El rango de los precios es: {rango_price}")
```

Interpretación: El rango de los precios es de 440.040, lo que indica que existe una amplia diferencia entre el computador de menor precio y el de mayor precio del conjunto de datos. Sin embargo, esta medida solo considera los valores extremos y no refleja cómo se distribuyen los datos entre ambos.

###6.2.2. Varianza
La varianza es una medida de dispersión que indica qué tan alejados se encuentran los datos respecto de su media. Se calcula considerando la diferencia entre cada valor y el promedio, elevando dichas diferencias al cuadrado y obteniendo posteriormente su promedio.

En este análisis se calculará la varianza del precio de los computadores portátiles para evaluar el grado de dispersión de sus valores.

```python
# Calculando la varianza con Pandas
varianza_precio = round(df["price"].var(), 2)

print(f"La varianza del precio de los computadores es: {varianza_precio}")
```

Interpretación: La varianza obtenida es elevada debido a que esta medida utiliza las diferencias respecto de la media elevadas al cuadrado. Esto indica que existe variabilidad en los precios de los computadores; sin embargo, al expresarse en unidades al cuadrado, su interpretación directa no resulta intuitiva. Por esta razón, a continuación se calculará la desviación estándar, que permitirá analizar la dispersión utilizando las mismas unidades del precio.

### 6.2.3 Desviación estándar

La desviación estándar es una medida de dispersión que indica, en promedio, cuánto se alejan los datos respecto de la media. A diferencia de la varianza, esta medida se expresa en las mismas unidades de la variable analizada, lo que facilita su interpretación.

En este análisis se calculará la desviación estándar del precio de los computadores portátiles para conocer el grado de variabilidad existente en el conjunto de datos.

```python
# Calculando la desviación estándar con Pandas
desviacion_precio = round(df["price"].std(), 2)

print(f"La desviación estándar del precio de los computadores es: {desviacion_precio}")
```

**Interpretación:** La desviación estándar del precio de los computadores es de **60.880,04**. Esto significa que, en promedio, los precios se alejan aproximadamente 60.880 unidades monetarias respecto de la media. Este resultado evidencia una alta dispersión en los precios del conjunto de datos, indicando la presencia de computadores con valores considerablemente diferentes entre sí.

La desviación estándar permite comprender de manera más intuitiva la dispersión de los datos que la varianza. Sin embargo, para determinar si esta variabilidad es alta o baja en relación con el promedio, es necesario calcular el **coeficiente de variación**, el cual será analizado en la siguiente sección.

### 6.2.4 Coeficiente de variación

El coeficiente de variación es una medida de dispersión relativa que expresa la desviación estándar como un porcentaje de la media. Permite comparar la variabilidad de un conjunto de datos independientemente de la unidad de medida.

En este análisis se calculará el coeficiente de variación del precio de los computadores portátiles para evaluar si los datos presentan un comportamiento homogéneo o heterogéneo.

```python
# Calculando el coeficiente de variación

media_precio = df["price"].mean()
desviacion_precio = df["price"].std()

coef_variacion = round((desviacion_precio / media_precio) * 100, 2)

print(f"El coeficiente de variación es: {coef_variacion}%")
```

**Interpretación:** El coeficiente de variación obtenido es de **76,19%**, lo que indica una alta variabilidad en los precios de los computadores portátiles. De acuerdo con este resultado, el conjunto de datos puede considerarse **heterogéneo**, ya que los precios presentan una dispersión importante respecto de la media.

| Medida                   |            Valor |
| ------------------------ | ---------------: |
| Rango                    |          440.040 |
| Varianza                 | 3.706.379.735,90 |
| Desviación estándar      |        60.880,04 |
| Coeficiente de variación |          76,19 % |

Las medidas de dispersión muestran que los precios de los computadores portátiles presentan una amplia variabilidad. El elevado coeficiente de variación (76,19%) indica que el conjunto de datos es heterogéneo, lo que significa que existen diferencias importantes entre los precios registrados. Este comportamiento coincide con lo observado anteriormente en las medidas de tendencia central, donde la media resultó superior a la mediana y la moda, sugiriendo la presencia de computadores de alto valor que incrementan el promedio del conjunto de datos.


##6.3 Medidas de posición
Las medidas de posición permiten dividir un conjunto de datos ordenados en partes iguales, facilitando la identificación de la ubicación relativa de las observaciones. Estas medidas complementan el análisis descriptivo al mostrar cómo se distribuyen los datos dentro del conjunto.

En esta sección se calcularán los cuartiles y algunos percentiles del precio de los computadores portátiles.

###6.3.1. Cuartiles
Los cuartiles son medidas de posición que dividen un conjunto de datos ordenados en cuatro partes iguales. Permiten identificar cómo se distribuyen los valores dentro del conjunto de datos y conocer la ubicación relativa de las observaciones.

```python
# Calculando los cuartiles

cuartiles = df["price"].quantile([0.25,0.50,0.75])

print(cuartiles)
```

**Interpretación:** Los cuartiles obtenidos muestran que el 25% de los computadores tienen un precio igual o inferior a **44.500** (Q1), el 50% presentan un precio igual o inferior a **61.990** (Q2 o mediana) y el 75% poseen un precio igual o inferior a **90.990** (Q3). Estos resultados permiten apreciar cómo se distribuyen los precios dentro del conjunto de datos.

###6.3.2. Percentiles
Los percentiles son medidas de posición que dividen un conjunto de datos ordenados en cien partes iguales. Permiten conocer el porcentaje de observaciones que se encuentran por debajo de un determinado valor y son ampliamente utilizados para analizar la distribución de los datos.

```python
# Calculando algunos percentiles

percentiles = df["price"].quantile([0.10,0.25,0.50,0.75,0.90])

print(percentiles)
```

**Interpretación:** Los percentiles permiten identificar la posición relativa de los precios dentro del conjunto de datos. Por ejemplo, el percentil 10 indica que el 10% de los computadores tienen un precio igual o inferior a **33.517,2**, mientras que el percentil 90 muestra que el 90% presentan un precio igual o inferior a **144.990**. Esto evidencia que la mayoría de los computadores se concentra bajo ese valor, existiendo un porcentaje reducido de equipos considerablemente más costosos.

## 7. Visualizaciones

Las visualizaciones permiten representar gráficamente la distribución y el comportamiento de los datos, facilitando su interpretación y complementando los resultados obtenidos mediante las medidas de estadística descriptiva.

### 7.1 Histograma del precio

El histograma muestra cómo se distribuyen los precios de los computadores portátiles dentro del conjunto de datos.

```python
plt.figure(figsize=(8,5))

plt.hist(df["price"], bins=20)

plt.title("Distribución del precio de los computadores")
plt.xlabel("Precio")
plt.ylabel("Frecuencia")

plt.show()
```

**Interpretación**
Se observa que la mayor concentración de computadores se encuentra en los rangos de precios bajos y medios, mientras que existe un número reducido de equipos de precio elevado. Esto explica que la media sea superior a la mediana.

### 7.2 Diagrama de caja (Boxplot)

El diagrama de caja permite visualizar la dispersión de los datos y detectar posibles valores atípicos.

```python
plt.figure(figsize=(8,2))

plt.boxplot(df["price"], vert=False)

plt.title("Boxplot del precio")

plt.show()
```

**Interpretación**
El gráfico evidencia la presencia de valores atípicos correspondientes a computadores de precios elevados. Estos valores incrementan el promedio del conjunto de datos y explican la diferencia observada entre la media y la mediana.

**Comparación del precio por marca mediante un diagrama de caja (Boxplot)**

El diagrama de caja permite comparar la distribución de los precios entre las principales marcas de computadores portátiles. Además de visualizar la mediana y la dispersión de los datos, facilita la identificación de posibles valores atípicos y diferencias en el comportamiento de los precios entre cada marca.

```python
# Seleccionar las 10 marcas con mayor cantidad de registros
marcas_principales = df["brand"].value_counts().head(10).index

df_marcas = df[df["brand"].isin(marcas_principales)]

plt.figure(figsize=(12, 6))

sns.boxplot(
    data=df_marcas,
    x="brand",
    y="price"
)

plt.title("Distribución del precio de las 10 marcas más frecuentes")
plt.xlabel("Marca")
plt.ylabel("Precio")

plt.xticks(rotation=45)
plt.tight_layout()

plt.show()
```

Interpretación: El diagrama de caja muestra que la distribución de los precios varía entre las distintas marcas de computadores. Algunas presentan una mayor dispersión de precios y una mediana más elevada, mientras que otras concentran sus precios en rangos más reducidos. Asimismo, se observan valores atípicos en varias marcas, lo que indica la existencia de modelos considerablemente más costosos que el resto.
